```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sqlite3
import seaborn as sns

sns.set(style='white', palette="Set2", color_codes=False)
sns.set_style("ticks")

%matplotlib inline
```

```python
db_path = 'data.db'
db = sqlite3.connect(db_path)
table_name = 'trail50-2024-03-12_15-58-53'
```

```python
def calculate_overhead_and_sleep_time(task):
    if task['tag'] == 'noop':
        task['sleep_time'] = 1e-3
        task['overhead'] = task['completion time']
    elif task['tag'] == 'sleep10ms':
        task['sleep_time'] = 0.01
        task['overhead'] = task['completion time'] - task['sleep_time']
    elif task['tag'] == 'sleep100ms':
        task['sleep_time'] = 0.1
        task['overhead'] = task['completion time'] - task['sleep_time']
    elif task['tag'] == 'sleep1000ms':
        task['sleep_time'] = 1
        task['overhead'] = task['completion time'] - task['sleep_time']
    elif task['tag'] == 'sleep10s':
        task['sleep_time'] = 10
        task['overhead'] = task['completion time'] - task['sleep_time']
    elif task['tag'] == 'sleep100s':
        task['sleep_time'] = 100
        task['overhead'] = task['completion time'] - task['sleep_time']
    return task

tasks = pd.read_sql_query(f'select * from "{table_name}"', db)
tasks['completion time'] = tasks['returned'] - tasks['start_submit']
tasks = tasks.apply(calculate_overhead_and_sleep_time, axis=1)
```

```python
data = pd.read_sql_query(f'select * from "{table_name}" where tag="sleep1000ms"', db)
data['completion time'] = data['returned'] - data['start_submit']
data = data.apply(calculate_overhead_and_sleep_time, axis=1)
```

```python
# 统计data中overhead的分布
sns.distplot(data['overhead'], kde=False, rug=True)
plt.title('Overhead Distribution')
plt.xlabel('Overhead (s)')
plt.ylabel('Count')
plt.show()
```

```python

print('Number of tasks with overhead < 0.15s:', len(data[data['overhead'] < 0.15]))
```

```python
monitor_tags = ['no_monitor', 'htex', 'diaspora']

tasks_per_trial = 3

fig, ax = plt.subplots()
cnt = 0

for monitor_tag in monitor_tags:
    cnt += 1
    df = tasks[(tasks['monitor_tag'] == monitor_tag)]
    # df = df.apply(lambda x: x.sort_values().iloc[1:-1])
    df = df.groupby('sleep_time', as_index=False)['overhead']
    means = df.mean()
    print(means)
    standard_deviations = df.std()
    means.plot('sleep_time', 'overhead', marker='o', ms=5, logx=True, yerr=list(standard_deviations['overhead']), label=monitor_tag, ax=ax)
    # if cnt == 1:
    #     break

# plt.ylim(0.1)
plt.ylabel('overhead [s]'.format(tasks_per_trial))
plt.xlabel('sleep time [s]')
# plt.savefig(f'{table_name}.pdf')
plt.legend(frameon=False)
```
