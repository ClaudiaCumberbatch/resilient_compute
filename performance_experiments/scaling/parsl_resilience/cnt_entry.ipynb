```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('data.db')
table1 = "trail2-2024-03-26_15-21-32"
table2 = "entry_cnt"
cmd = f'''
SELECT "{table1}".run_id,monitor_tag,workers,tasks_per_trial,trial,"{table1}".app_name,record_num FROM "{table1}"
JOIN "{table2}"
ON '{table1}'.run_id = '{table2}'.run_id
'''
tasks = pd.read_sql_query(cmd, conn)
tasks
```
